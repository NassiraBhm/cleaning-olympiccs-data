# Clean athletes data:

#### Get rid of bullet points in Used Names

```python
data['Name'] = data['Used name'].str.replace("•", " ")
```

#### Split height/weight

```python
data[['height_cm', 'weight_kg']] = data['Measurements'].str.split('/', expand=True)
```

```python
data['height_cm'] = pd.to_numeric(data['height_cm'].str.strip(' cm'), errors='coerce')
data['weight_kg'] = pd.to_numeric(data['weight_kg'].str.strip(' kg'), errors='coerce')
```

```python
data.head()
```

#### Parse out dates from Born & Died column

```python
datePattern = r'(\d+ \w+ \d{4}|\d{4})'
data['birth_date'] = data['Born'].str.extract(datePattern)
data['birth_date'] = pd.to_datetime(data['birth_date'], format='mixed', errors='coerce')
```

```python
data['death_date'] = data['Died'].str.extract(datePattern)
data['death_date'] = pd.to_datetime(data['death_date'], format='mixed', errors='coerce')
```

```python
data.head()
```

#### Parse out city, region, and country from Born column

```python
locationPattern = r'in ([\w\s()-]+), ([\w\s-]+) \((\w+)\)'
data[['birth_city', 'birth_region', 'birth_country']] = data['Born'].str.extract(locationPattern)
```

```python
pd.set_option('display.max_colwidth', None)
data.head()
```

#### Get rid of extra columns

```python
CleanData = data[['athlete_id', 'Name', 'birth_date', 'birth_city', 'birth_region',
                   'birth_country', 'NOC', 'height_cm', 'weight_kg', 'death_date']]
```

```python
CleanData.head()
```

```python
CleanData.to_csv('CleanData',index = False)
```

# Clean results data:

```python
results = pd.read_csv('results.csv')
results.head()
```

```python
df = results.copy()
```

#### Split year and type from games

```python
#yearPattern = r'(\d{4})'
typePattern = r"\d+ (\w+ \w+)"
df ['year'] = df['Games'].str.extract(yearPattern)
df['type'] = df['Games'].str.extract(typePattern)
```

```python
df ['year'] = pd.to_numeric(df ['year'])
```

```python
df.head()
```

#### split out tied column from position

```python
df['place'] = df['Pos'].str.extract(r'(\d+)')
df['tied'] = df['Pos'].str.contains('=')
```

```python
df['place'] = pd.to_numeric(df['place'])
df['tied'] = df['tied'].astype(bool)
```
